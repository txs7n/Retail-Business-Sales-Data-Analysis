## Data Cleaning

```python
# we see that all the TransactionNo starts with a C, and the quantity are negative
# which means the data might require some cleaning
# first we examine the phenomenon to see the occurence within the dataset
```

```python
df['TransactionNo'].value_counts()
```

```python
# to check how many entries in TransactionNo starts with the letter 'C'

df.query("TransactionNo.str.startswith('C')")
```

```python
# to check how many entries in TransactionNo do not start with the letter 'C'

df.query("~TransactionNo.str.startswith('C')")
```

```python
# in essence, 1.6% of the entire dataset starts has transaction numbers starting with 'C'
# safe to clean this up on the assumption that it could have been an error from the machine at point of sale
# assumption can be somewhat corroborated by the fact the TransactionNo that starts with 'C' have negative Quantity values
# as observed from the investigation below:
```

```python
# to check Quantity values with negative sign

df.loc[df['Quantity'] < 0].sort_values(['Quantity']).head(20)
```

```python
# picking an arbitrary Quantity value to check for its +ve and -ve occurence within the table

df.loc[(df['Quantity'] == 960 ) | (df['Quantity'] == -960)]
```

```python
df.loc[(df['Quantity'] == 720 ) | (df['Quantity'] == -720)]
```

```python
# both Quantity numbers (960 & 720) corroborated the assumption
# since their -ve occurence in the dataframe corresponds to a TransactionNo variable that starts with a 'C'
# with the assumption theory satisfied, we begin the cleaning process
```

```python
# first, make a copy of the initial dataset

df1 = df.copy()
```

```python
df1
```

```python
# Create a function to remove the -ve sign for all negative Quantity variables 

def neg_replace(x):
    if x < 0:
        return -1*x
    else:
        return x        
```

```python
# applying the function on the Quantity column

df1['Quantity'] = df1['Quantity'].apply(neg_replace)
```

```python
# crosschecking 

df1.loc[df1['Quantity'] < 0]
```

```python
# to clean up the TransactionNo column

df1.loc[df1['TransactionNo'].str.startswith('C')]
```

```python
df1['TransactionNo'] = df1['TransactionNo'].str.lstrip('C')
```

```python
df1.loc[df1['TransactionNo'].str.startswith('C')]
```

```python
df1.loc[df1.duplicated()].sort_values(['Quantity'], ascending = False)
```

```python
# Next, to check for duplicate values

df1.loc[df1.duplicated()]
```

```python
# investigating duplicate values

df1.loc[(df1['ProductName'] == 'Set Of 3 Hanging Owls Ollie Beak') & (df1['TransactionNo'] == '581538')]
```

```python
# from the looks of it, the duplication might have been an error, not intentional.
# we can then drop these duplicate values
```

```python
# dropping the duplicate values and assigning to a new dataframe

df2 = df1.drop_duplicates().copy()
```

## Feature Engineering

```python
# we need to create a couple of features for our dataset to aid in our purpose of analysis
```

```python
# first we create a TotalAmt column which is the multiplication of the Price and Quantity columns

df2['TotalAmt'] = df2['Price']*df2['Quantity']
```

```python
df2
```

```python
# Next, we create a Month column by extracting the month name from our Date column

df2['Month'] = df2['Date'].dt.month_name()
```

## Visualizing the Data (For Stakeholders)

```python
import matplotlib.ticker as ticker

# creating a df that will best suit the horizontal bar chart
df_most_prod_sold = df2.groupby('ProductName')[['TotalAmt']].sum().sort_values(['TotalAmt'], ascending = True).tail(10).reset_index()



axes = df_most_prod_sold.reset_index().plot(kind='barh', x='ProductName', y='TotalAmt', figsize=(15,11))

# Use a FuncFormatter to format x-axis ticks
formatter = ticker.FuncFormatter(lambda x, pos: '{:,.1f}M'.format(x*1e-6))
axes.xaxis.set_major_formatter(formatter)

axes.tick_params(axis='x', labelsize=14) 
axes.tick_params(axis='y', labelsize=14)

plt.xlabel('Total Amount (in million USD)', fontsize = 25, labelpad = 20)
plt.ylabel('Product Name', fontsize = 25, labelpad = 20)
plt.title('Total Amount Sold per Product', fontsize = 35, y = 1.05)

plt.savefig('total_amt_sold_per_product.png', bbox_inches = 'tight')

plt.show()
```

```python
# creating a df that will best suit the horizontal bar chart
df_totalamt_months = df2.groupby(['Month'])[['TotalAmt']].sum().sort_values(['TotalAmt'], ascending = True)



axes_1 = df_totalamt_months.plot(kind='barh', figsize=(15,10))

formatter = ticker.FuncFormatter(lambda x, pos: '{:,.1f}M'.format(x*1e-6))
axes_1.xaxis.set_major_formatter(formatter)

axes_1.tick_params(axis='x', labelsize=14) 
axes_1.tick_params(axis='y', labelsize=14)

plt.xlabel('Total Amount (in million USD)', fontsize=25, labelpad = 20)
plt.ylabel('Month', fontsize=25, labelpad=20)
plt.title('Total Amount sold per Month', fontsize=30, y = 1.05)

plt.savefig('total_amt_sold_per_month.png', bbox_inches = 'tight')

plt.show()
```

```python
# creating a df that will best suit the structure of the horizontal bar chart
df_totalamt_country = df2.groupby(['Country'])[['TotalAmt']].sum().sort_values(['TotalAmt'], ascending = True).tail(10).reset_index()

axes_2 = df_totalamt_country.plot(kind = 'barh', x = 'Country', y = 'TotalAmt', figsize = (15,10))

formatter = ticker.FuncFormatter(lambda x, pos: '{:,.1f}M'.format(x*1e-6))
axes_2.xaxis.set_major_formatter(formatter)

axes_2.tick_params(axis='x', labelsize=14) 
axes_2.tick_params(axis='y', labelsize=14)

plt.xlabel('Total Amount (in million USD)', fontsize=25, labelpad = 20)
plt.ylabel('Country', fontsize=25, labelpad = 20)
plt.title('Total Amount sold per Country', fontsize=30, y = 1.05)

plt.savefig('total_amt_sold_per_country.png', bbox_inches = 'tight')

plt.show()
```

```python
# creating a df that will best suit the structure of the horizontal bar chart
df_most_freq_sold_prod = df2['ProductName'].value_counts().reset_index().rename(columns={'index':'Product Name', 'ProductName':'Count'}).sort_values('Count', ascending = True).tail(10).reset_index()

df_most_freq_sold_prod.drop('index', axis = 1, inplace = True)



axes_3 = df_most_freq_sold_prod.plot(kind = 'barh', x = 'Product Name', y = 'Count', figsize = (15,11))

axes_3.tick_params(axis='x', labelsize=14) 
axes_3.tick_params(axis='y', labelsize=14)

plt.xlabel('No of Times Sold', fontsize = 25, labelpad = 20)
plt.ylabel('Product Name', fontsize = 25, labelpad = 20)
plt.title('Most Frequently Sold Products', fontsize = 30, y = 1.05)

plt.savefig('most_freq_sold_products.png', bbox_inches = 'tight')

plt.show()
```
